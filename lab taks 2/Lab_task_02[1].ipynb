```python


start = 1
end = 100

for i in range(start, end + 1):
    word = ""

    if i % 3 == 0:
        word = word + "Fizz"

    if i % 5 == 0:
        word = word + "Buzz"

    if word == "":
        print(i)
    else:
        print(word)
```

```python
movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]

extra_movies = int(input("How many extra movies? "))

for i in range(extra_movies):
    name = input("Movie name: ")
    cost = int(input("Budget: "))
    movies.append((name, cost))

total_budget = 0
for m in movies:
    total_budget = total_budget + m[1]

movie_count = len(movies)
average = total_budget / movie_count
print("\nAverage budget is", average)

above = 0
print("\nMovies above average:\n")

for m in movies:
    title = m[0]
    budget = m[1]
    if budget > average:
        extra = budget - average
        print(title, "is", extra, "over average")
        above = above + 1

print("\nNumber of movies above average =", above)
```
