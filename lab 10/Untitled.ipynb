```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data(1).csv')
print(data.head())
```

```python
print('the rows in this is  {} rows.'.format(data.shape[0]))
print('the columns in this is  {} columns.'.format(data.shape[1]))
```

```python
print(np.sum(pd.isnull(data)))
```

```python
print(data['statezip'].unique())
```

```python
num = data['statezip'].mode()[0]
data['statezip'] = data['statezip'].fillna(num)
```

```python
data['bedrooms'] = data['bedrooms'].astype(np.int64)
```

```python
data.drop('view', axis=1, inplace=True)
```
